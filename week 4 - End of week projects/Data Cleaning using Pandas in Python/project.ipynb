```python
# libraries
import pandas as pd
import numpy as np
```

```python
#read the dataset :
df = pd.read_excel("Customer_Call_List_Maroc.xlsx")
print('the dataset :\n')
df
```

```python
#indicating the dataset size before cleaning :
print('the dataset before cleaning :')
print(f"the dataset's number of raws is : {df.shape[0]}")
print(f"the dataset's number of columns is : {df.shape[1]}")
```

```python
#dataset showcasing duplicates
df[df.duplicated(subset=['First_Name', 'Last_Name', 'Phone_Number'])]
```

```python
#droping duplicates :
df.drop_duplicates(subset=['First_Name', 'Last_Name', 'Phone_Number'], inplace = True)
df
```

```python
#indicating the dataset size after cleaning :
print('the dataset after cleaning :')
print(f"the dataset's number of raws is : {df.shape[0]}")
print(f"the dataset's number of columns is : {df.shape[1]}")
```

```python
#indentification and remove of unusefull columns :
#indentifcation methodologie followed :
#-- if the first and last name are not exists\
#-- if phone number or adress not exist or not valid, since our mine contact bridge with the csutomer is phone number or its location atlest

# 1. Drop if both names are missing
df = df.dropna(subset=['First_Name', 'Last_Name'], how='all')

# 2. keep only names greater than 4 characters
df = df[ (df['First_Name'].str.len() >= 3) & (df['Last_Name'].str.len() >= 3) ]

# 3. Drop if Phone or Address is missing
df = df.dropna(subset=['Phone_Number', 'Address'], how='any')


print("Cleaning complete. Data remaining:")
df
```

```python
#indicating the dataset size after more cleaning :
print('the dataset after second cleaning :')
print(f"the dataset's number of raws is : {df.shape[0]}")
print(f"the dataset's number of columns is : {df.shape[1]}")
```

```python
# drop the Not_Useful_Column :
df = df.drop(columns=["Not_Useful_Column"])
df
```

```python
#data cleaning from non wanted charachters in each column :
#data cleaning for the First_Name column :
df["First_Name"]= df["First_Name"].str.replace(r'[^A-Za-z ]', '', regex=True)

#data cleaning for the Last_Name column :
df["Last_Name"]= df["Last_Name"].str.replace(r'[^A-Za-z ]', '', regex=True)

#data cleaning for the Phone_Number column :
df["Phone_Number"]= df["Phone_Number"].str.replace(r'[^\d]', '', regex=True)
```

```python
print('dataset after cleaning from non wanted charachters :')
df
```

```python
#formaing Phone Number into 3-3-4 format 
df['Phone_Number'] = df['Phone_Number'].str[:3]+'-'+df['Phone_Number'].str[3:6]+'-'+df['Phone_Number'].str[6:]

print("dataset after formatting phone number :")
df
```

```python
#split the address column into 3 new columns: street address, state, zip code
df['Street_Address'] = df['Address'].str.split(',').str[0]
df['State'] = df['Address'].str.split(',').str[1]
df['Zip_Code'] = df['Address'].str.split(',').str[2]

print("dataset after Address column split :")
df
```

```python
# drop the Address column
df.drop('Address', axis=1, inplace=True)
df
```

```python
#identifying states for zipcode alignement :
State = df.drop_duplicates(subset="State")
State["State"]
```

```python
#zip code assignement
zip_codes = {'casablanca' : 20000, 'rabat': 10000, 'fez' : 30000, 'fes' : 30000, 'fès' : 30000}
df["State"] = df["State"].str.lower().str.strip()
df["Zip_Code"] = df["State"].map(zip_codes)

print('dataset after zip code assignement : \n')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Y", " Do not contact").str.replace("N", "Contact authorized")

print("dataset :\n")
df
```
