## 1. Charger le CSV brut avec Pandas (london_merged.csv):

```python
df = pd.read_csv("london_merged.csv")
print('dataset : \n')
df
```

## 2. Observer le dataset :

### 2.1. Nombre de lignes et colonnes

```python
# check for the size of the dataset :
print(f'dataset size :\n')
print(f"dataset's number of raws : {df.shape[0]}")
print(f"dataset's number of columns : {df.shape[1]}")
```

### 2.2. Types de variables :

```python
#check for data type of each column 
print('data type of each column :')
df.dtypes
```

### 2.3. Vérifier les valeurs manquantes :

```python
#check all rows with atleast one value missing :
df_isnull = df[df.isnull().any(axis = 1)]
print(f"Number of rows with at least one missing value : {df_isnull.shape[0]} rows")
print(f'Numbers of missing values in each column : \n{df_isnull.sum()}')
```

### 2.4. Vérifier les Doublons:

```python
#check for duplicates
duplicated = df[df.duplicated()]
print(f"Number of duplicated rows: {duplicated.shape[0]}")
```

## 3. Renommer les colonnes pour les rendre compréhensibles :

```python
#renaming columns :
dict = {"cnt" : "Nombre de trajets", "t1" : "Température réelle (°C)","t2" : "Température ressentie (°C)", "hum" : "Humidité", "wind_speed" : "Vitesse du vent (km/h)", "weather_code" : "Météo", "season" : "Saison"}
df.rename(columns=dict, inplace=True)
df
```

## 4. Conversion et nettoyage :

### 4.1. conversion l'echelle des variables humidité de pourcentage (%) en décimal (entre 0 et 1) :

```python
#replace humidity values : from % scale to decimal scale :
df['Humidité'] = df['Humidité'] / 100
df
```

```python
#note from ssi omar 
df.describe()
```

### 4.2. Conversion des codes numériques en valeurs explicites pour les variables "Saison" et "Meteo":

```python
#Map season conditions
dict_saison = {3 : 'Hiver', 2: 'Automne', 0 : 'Printemps', 1: 'Ete'}
df['Saison'] = df['Saison'].replace(dict_saison)
# Map weather conditions
Meteo = {1 : 'Clair', 2 : 'Nuages épars', 3 : 'Nuages fragmentés', 4 : 'Couvert', 7 : 'Pluie', 10 : 'Pluie avec orage', 26 : 'Neige'} 
df['Météo'] = df['Météo'].replace(Meteo)
df
```

## 6. Export du dataset clean (london_bikes_final.xlsx ou .csv)

```python
df.to_excel("london_merged_after_preprocessing_pandas.xlsx", index=False)
print("Dataframe saved to excel file!")
```
