# Marketing Campaign Analysis

## Import Libraries and Load Dataset
We begin by loading the relevant libraries and our marketing dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Formatting settings
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load dataset
df = pd.read_csv('dataset.csv')
df.columns = df.columns.str.strip() # Remove spaces in column names
df['Income'] = df['Income'].fillna(df['Income'].median()) # Handle NaNs explicitly
df.head()
```

## Phase 1: Data Preparation & Feature Engineering
In this phase, we add new columns that summarize spending, demographics, and campaign interactions to make our visualizations cleaner.

```python
# Customer Age
current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth']

# Total Spending across all product categories
df['Total_Spending'] = (df['MntWines'] + df['MntFruits'] + 
                        df['MntMeatProducts'] + df['MntFishProducts'] + 
                        df['MntSweetProducts'] + df['MntBakedProds'])

# Total Dependents
df['Total_Dependents'] = df['Kidhome'] + df['Teenhome']

# Total Campaigns Accepted
df['Total_Campaigns_Accepted'] = (df['AcceptedCmp1'] + df['AcceptedCmp2'] + 
                                  df['AcceptedCmp3'] + df['AcceptedCmp4'] + 
                                  df['AcceptedCmp5'] + df['Response'])

# Customer Tenure (Days)
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%m/%d/%Y', errors='coerce')
# For a static dataset, let's use the max date in the dataset as 'today' to calculate tenure accurately
last_date = df['Dt_Customer'].max()
df['Customer_Tenure_Days'] = (last_date - df['Dt_Customer']).dt.days

df[['Age', 'Total_Spending', 'Total_Dependents', 'Total_Campaigns_Accepted', 'Customer_Tenure_Days']].head()
```

## Phase 2A: Demographics Profiling
Who are our customers? Let's look at income, age, and marital status distributions.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Income by Education
sns.boxplot(ax=axes[0], data=df, x='Education', y='Income', palette='Set2')
axes[0].set_title('Income by Education Level')

# Age Distribution
sns.histplot(ax=axes[1], data=df, x='Age', bins=20, kde=True, color='purple')
axes[1].set_title('Age Distribution')

# Marital Status
sns.countplot(ax=axes[2], data=df, x='Marital_Status', palette='Set3', 
              order=df['Marital_Status'].value_counts().index)
axes[2].set_title('Marital Status Breakdown')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## Phase 2B: Product Purchasing Habits
What do they buy? Which categories generate the most revenue?

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Revenue by Category
categories = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntBakedProds']
category_sums = df[categories].sum().sort_values(ascending=False)

sns.barplot(ax=axes[0], x=category_sums.values, y=category_sums.index, palette='viridis')
axes[0].set_title('Total Revenue by Product Category')
axes[0].set_xlabel('Total Revenue')
axes[0].set_ylabel('Categories')

# Income vs Total Spending
sns.scatterplot(ax=axes[1], data=df, x='Income', y='Total_Spending', alpha=0.5, color='teal')
axes[1].set_title('Income vs Total Spending')

plt.tight_layout()
plt.show()
```

## Phase 3A: Evaluating Marketing Channels
Where do our customers make their purchases? Web, Catalog, or Store?

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Total purchases by channel
channels = ['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
channels_sums = df[channels].sum().sort_values(ascending=False)

sns.barplot(ax=axes[0], x=channels_sums.index, y=channels_sums.values, palette='magma')
axes[0].set_title('Purchases by Sales Channel')
axes[0].set_ylabel('Total Purchases')

# Web Visits vs Web Purchases
sns.scatterplot(ax=axes[1], data=df, x='NumWebVisitsMonth', y='NumWebPurchases', alpha=0.5, color='orange')
axes[1].set_title('Web Visits vs Web Purchases per Month')

plt.tight_layout()
plt.show()
```

## Phase 3B: Campaign Success
How effective were our ad campaigns, and who accepted them?

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Acceptance rate by campaign
campaigns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']
campaigns_sums = df[campaigns].sum() / len(df) * 100 # % Conversion Rate

sns.barplot(ax=axes[0], x=campaigns_sums.index, y=campaigns_sums.values, palette='coolwarm')
axes[0].set_title('Campaign Conversion Rates (%)')
axes[0].tick_params(axis='x', rotation=45)

# Profile of Campaign 1 Acceptors (Income)
sns.boxplot(ax=axes[1], data=df, x='AcceptedCmp1', y='Income', palette='Pastel1')
axes[1].set_title('Income Profile of Campaign 1 Responders')
axes[1].set_xticklabels(['Rejected', 'Accepted'])

plt.tight_layout()
plt.show()
```
