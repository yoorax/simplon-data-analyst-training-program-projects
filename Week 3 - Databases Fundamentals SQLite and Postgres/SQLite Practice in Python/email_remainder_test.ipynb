```python
import sqlite3 as sq
import pandas as pd

DB_connect = sq.connect('bibliotheque.db')
cursor = DB_connect.cursor()
cursor.execute("PRAGMA foreign_keys=ON;")
print('database bibliotheque has been connected seccussfully!')
```

```python
query_table_test_emprunts_creation ="""CREATE TABLE IF NOT EXISTS emprunts_test(
            emprunt_id INTEGER PRIMARY KEY AUTOINCREMENT,
            livre_id INTEGER REFERENCES livres(livre_id) ON DELETE CASCADE,
            emprunteur_id INTEGER REFERENCES emprunteurs(emprunteur_id) ON DELETE CASCADE,
            date_emprunt DATE NOT NULL,
            date_retour_prevue DATE NOT NULL,       
            date_retour_effective DATE);"""
try :
    cursor.execute(query_table_test_emprunts_creation)
    print("table emprunts_test has been created successfully!")
except Exception as e:
    print(f"Error : {e}")
```

```python
cursor.execute(" SELECT * FROM emprunts_test;")
emprunts_test = cursor.fetchall()

df_emprunts_test = pd.DataFrame(emprunts_test, columns = ['emprunt_id','livre_id', 'emprunteur_id', 'date_emprunt', 'date_retour_prevue', 'date_retour_effective'])
print(f"emprunts_test table : \n\n{df_emprunts_test}")
```

```python
# Updated SQL Query to include the email address
query_join_with_email = """
SELECT 
    p.prenom, 
    p.nom, 
    p.email, 
    l.titre, 
    e.date_emprunt, 
    e.date_retour_prevue, 
    e.date_retour_effective 
FROM emprunts_test e
JOIN emprunteurs p ON e.emprunteur_id = p.emprunteur_id
JOIN livres l ON e.livre_id = l.livre_id;
"""

# Execute and fetch results
cursor.execute(query_join_with_email)
data_detailed = cursor.fetchall()

# Create DataFrame with the requested columns
df_detailed = pd.DataFrame(data_detailed, columns=[
    'Prenom', 
    'Nom', 
    'Email', 
    'Livre', 
    'Date Emprunt', 
    'Date Retour Prévue', 
    'Date Retour Effective'
])

# Display the result
print("Detailed Loans Table with Emails:")
df_detailed
```
