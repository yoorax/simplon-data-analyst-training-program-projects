#Create database and connect it :

```python
#import sqlite3 module :
import sqlite3

try :
    # connect to database (and create it if it doesn't exist)
    DB_connection = sqlite3.connect('bibliotheque.db')

    # create a cursor object :
    cursor = DB_connection.cursor()

    # enable (Activate) foreign keys :
    cursor.execute("PRAGMA foreign_keys = ON;")

    #show tables :
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    print(cursor.fetchall())

except sqlite3.Error as e :
    print("Error : connection/creation of the database failed : ", e)

finally :
    if DB_connection :
        DB_connection.close()
        print("connection closed")
```
